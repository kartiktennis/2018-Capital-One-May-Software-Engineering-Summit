```python
coordinates = pd.read_csv("location_coordinates.csv")
coordinates
```

```python
df_SF = coordinates.loc[coordinates['call_type'] == 'Structure Fire']
df_AL = coordinates.loc[coordinates['call_type'] == 'Alarms']
df_MI = coordinates.loc[coordinates['call_type'] == 'Medical Incident']

value_list = ['Structure Fire', 'Alarms', 'Medical Incident']
df_MISC = coordinates[~coordinates.call_type.isin(value_list)]
```

```python
Structure_Fires = df_SF.to_csv("Structure_Fires.csv") 
Alarms = df_AL.to_csv("Alarms.csv") 
Medical_Incidents = df_MI.to_csv("Medical_Incident.csv") 
Miscellaneous = df_MISC.to_csv('Miscellaneous.csv')
```
