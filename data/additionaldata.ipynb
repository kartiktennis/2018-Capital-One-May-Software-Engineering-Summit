```python
zipcodes = pd.read_csv("zipcodes_csv.csv")
```

```python
from datetime import datetime,timedelta

dispatch_times = pd.read_csv('dispatch_times_cleansed.csv')

def createTimeDelta(x):
    time = datetime.strptime(x,'%D days %H:%M:%S')
    return timedelta(hours=time.hour,minutes=time.minute,seconds=time.second);

#zipcodes['dispatch_time'] = dispatch_times['dispatch_time']
zipcodes.zipcode_of_incident.unique()
```

```python
#structure fires, trains, water rescues take longest times for dispatch - get corresponding zipcodes
Structure_Fire_code = pd.DataFrame(columns=['call_type', 'zipcode_of_incident'])
Water_Rescue_code = pd.DataFrame(columns=['call_type', 'zipcode_of_incident'])
TrainRail_code = pd.DataFrame(columns=['call_type', 'zipcode_of_incident'])

for index,row in zipcodes.iterrows(): 
    if row['call_type'] == 'Structure Fire':
        Structure_Fire_code = zipcodes.loc[zipcodes['call_type'] == 'Structure Fire']
    if row['call_type'] == 'Water Rescue':
        Water_Rescue_code = zipcodes.loc[zipcodes['call_type'] == 'Water Rescue']
    if row['call_type'] == 'Train / Rail Incident':
        TrainRail_code = zipcodes.loc[zipcodes['call_type'] == 'Train / Rail Incident']

# Structure_Fire_code #SF(longest dispatch type) with zipcode areas
# Water_Rescue_code #WR(longest dispatch type) with zipcode areas
# TrainRail_code #TR(longest dispatch type) with zipcode areas

#get most common zipcode in each of three files - corresponds to the longest dispatch areas

zipcode_Structure_Fire = Structure_Fire_code['zipcode_of_incident'].mode()

zipcode_TrainRail = TrainRail_code['zipcode_of_incident'].mode()

zipcode_Water_Rescue = Water_Rescue_code['zipcode_of_incident'].mode()

print("Most common Structure Fire Area: " + str(zipcode_Structure_Fire))
print("Most common Train / Rail Incident Area: " + str(zipcode_TrainRail))
print("Most common Water Rescue Area: " + str(zipcode_Water_Rescue))
```
